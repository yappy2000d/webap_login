```
import numpy as np
import matplotlib.pyplot as plt


p1 = 0.8238
p2 = 0.9898

t1 = 2.45
t2 = 1.18 * t1

x = np.linspace(0, 200, 2000)

E1 = (x + t1) / p1
E2 = (x + t2) / p2

plt.plot(x, E2, label="EucDist")
plt.plot(x, E1, label="TFLite")

plt.fill_between(x, E1, E2, alpha=0.2)

plt.legend()

plt.title("Execpted Time vs. Image Fetch Time")

plt.xlabel("Image Fetch Time per Image (ms)")
plt.ylabel("Execpted Time per Image(ms)")
# plt.savefig("execpted_time.webp", dpi=200)
plt.show()
```

```
def improve(x, t1):
    return (1 - (E2/E1)) * 100

plt.title("Relative Improvement vs. Image Fetch Time")

plt.xlabel("Image Fetch Time (ms)")
plt.ylabel("Relative Speed Improvement (%)")

plt.plot(x, improve(x, t1))
plt.show()
# plt.savefig("improve.webp", dpi=200)
```
